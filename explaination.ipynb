```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image
image = cv2.imread('R.png')

# Apply smoothing using a box filter
kernel_size = (9, 9)  # Adjust the kernel size as needed
smoothed_image = cv2.boxFilter(image, -1, kernel_size)

# Convert images to grayscale
gray_original = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_smoothed = cv2.cvtColor(smoothed_image, cv2.COLOR_BGR2GRAY)

# Calculate histograms
hist_original = cv2.calcHist([gray_original], [0], None, [256], [0, 256])
hist_smoothed = cv2.calcHist([gray_smoothed], [0], None, [256], [0, 256])

# Plot histograms
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.plot(hist_original, color='black')
plt.title('Original Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(122)
plt.plot(hist_smoothed, color='black')
plt.title('Smoothed Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image
image = cv2.imread('R.png')

# Apply smoothing using a Gaussian filter
kernel_size = (9, 9)  # Adjust the kernel size as needed
sigmaX = 0  # Standard deviation in X direction (0 indicates auto-calculation based on kernel size)
smoothed_image = cv2.GaussianBlur(image, kernel_size, sigmaX)

# Convert images to grayscale
gray_original = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_smoothed = cv2.cvtColor(smoothed_image, cv2.COLOR_BGR2GRAY)

# Calculate histograms
hist_original = cv2.calcHist([gray_original], [0], None, [256], [0, 256])
hist_smoothed = cv2.calcHist([gray_smoothed], [0], None, [256], [0, 256])

# Plot histograms
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.plot(hist_original, color='black')
plt.title('Original Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(122)
plt.plot(hist_smoothed, color='black')
plt.title('Smoothed Image Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('R.png', cv2.IMREAD_GRAYSCALE)

# Apply box filter
box_filtered_image = cv2.blur(image, (5, 5))  # You can adjust the kernel size (5, 5) as needed

# Apply Gaussian filter
gaussian_filtered_image = cv2.GaussianBlur(image, (5, 5), 0)  # You can adjust the kernel size (5, 5) and sigma as needed

# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(box_filtered_image, cmap='gray')
plt.title('Box Filter')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(gaussian_filtered_image, cmap='gray')
plt.title('Gaussian Filter')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.hist(image.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(2, 3, 5)
plt.hist(box_filtered_image.ravel(), bins=256, color='red', alpha=0.7)
plt.title('Histogram after Box Filter')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(2, 3, 6)
plt.hist(gaussian_filtered_image.ravel(), bins=256, color='green', alpha=0.7)
plt.title('Histogram after Gaussian Filter')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('R.png', cv2.IMREAD_GRAYSCALE)

# Apply Gaussian filter
gaussian_filtered_image = cv2.GaussianBlur(image, (5, 5), 0)  # You can adjust the kernel size (5, 5) and sigma as needed

# Apply median filter
median_filtered_image = cv2.medianBlur(image, 5)  # You can adjust the kernel size (5) as needed

# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(gaussian_filtered_image, cmap='gray')
plt.title('Gaussian Filter')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(median_filtered_image, cmap='gray')
plt.title('Median Filter')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.hist(image.ravel(), bins=256, color='blue', alpha=0.7)
plt.title('Histogram of Original Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(2, 3, 5)
plt.hist(gaussian_filtered_image.ravel(), bins=256, color='red', alpha=0.7)
plt.title('Histogram after Gaussian Filter')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(2, 3, 6)
plt.hist(median_filtered_image.ravel(), bins=256, color='green', alpha=0.7)
plt.title('Histogram after Median Filter')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
